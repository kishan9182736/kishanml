```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('iris.csv')
df.head()
```

Scatter Plot

```python
Setosa = df[df['species'] == 'setosa'][['sepal_length', 'sepal_width']]
Versicolor = df[df['species'] == 'versicolor'][['sepal_length', 'sepal_width']]
Virginica = df[df['species'] == 'virginica'][['sepal_length', 'sepal_width']]
```

```python
markers = {'Setosa': '+', 'Versicolor': '^', 'Virginica': '*'}

plt.scatter(Setosa['sepal_length'], Setosa['sepal_width'], color = 'red', marker=markers['Setosa'], label = 'Setosa')
plt.scatter(Versicolor['sepal_length'], Versicolor['sepal_width'], color = 'blue', marker=markers['Versicolor'], label = 'Versicolor')
plt.scatter(Virginica['sepal_length'], Virginica['sepal_width'], color = 'green', marker=markers['Virginica'],  label = 'Virginica')

plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Scatter Plot of Sepal Length vs. Sepal Width by Species')
plt.legend(title='Species')
plt.show()
```

```python
Setosa = df[df['species'] == 'setosa'][['petal_length', 'petal_width']]
Versicolor = df[df['species'] == 'versicolor'][['petal_length', 'petal_width']]
Virginica = df[df['species'] == 'virginica'][['petal_length', 'petal_width']]
```

```python
markers = {'Setosa': '+', 'Versicolor': '^', 'Virginica': '*'}

plt.scatter(Setosa['petal_length'], Setosa['petal_width'], color = 'red', marker=markers['Setosa'], label='Setosa')
plt.scatter(Versicolor['petal_length'], Versicolor['petal_width'], color = 'blue', marker=markers['Versicolor'], label='Versicolor')
plt.scatter(Virginica['petal_length'], Virginica['petal_width'], color = 'green', marker=markers['Virginica'], label='Virginica')

plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Scatter Plot of Petal Length vs. Petal Width by Species')
plt.legend(title='species')
plt.show()
```

Box Plot

```python
Setosa = df[df['species'] == 'setosa']
Versicolor = df[df['species'] == 'versicolor']
Virginica = df[df['species'] == 'virginica']
```

```python
sepal_length_data = [Setosa['sepal_length'], Versicolor['sepal_length'], Virginica['sepal_length']]
sepal_width_data = [Setosa['sepal_width'], Versicolor['sepal_width'], Virginica['sepal_width']]
labels = ['Setosa', 'Versicolor', 'Virginica']

plt.boxplot(sepal_length_data, tick_labels=labels)
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.title('Box Plot of Sepal Length by Species')
plt.show()

plt.boxplot(sepal_width_data, tick_labels=labels)
plt.xlabel('Species')
plt.ylabel('Sepal Width (cm)')
plt.title('Box Plot of Sepal Width by Species')
plt.show()
```

```python
petal_length_data = [Setosa['petal_length'], Versicolor['petal_length'], Virginica['petal_length']]
petal_width_data = [Setosa['petal_width'], Versicolor['petal_width'], Virginica['petal_width']]
labels = ['Setosa', 'Versicolor', 'Virginica']

plt.boxplot(petal_length_data, tick_labels=labels)
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.title('Box Plot of Petal Length by Species')
plt.show()

plt.boxplot(petal_width_data, tick_labels=labels)
plt.xlabel('Species')
plt.ylabel('Petal Width (cm)')
plt.title('Box Plot of Petal Width by Species')
plt.show()
```

```python
Mean_Sepal_Length = df.groupby('species')['sepal_length'].mean()
Mean_Sepal_Width = df.groupby('species')['sepal_width'].mean()

summary_stats = pd.DataFrame({
    'Mean Sepal Length': Mean_Sepal_Length,
    'Mean Sepal Width': Mean_Sepal_Width
}).reset_index()
```

```python
labels = summary_stats['species']
Mean_Sepal_Length = summary_stats['Mean Sepal Length']
Mean_Sepal_Width = summary_stats['Mean Sepal Width']

plt.bar(labels, Mean_Sepal_Length, color='skyblue')
plt.xlabel('Species')
plt.ylabel('Mean Sepal Length (cm)')
plt.title('Bar Plot of Mean Sepal Length by Species')
plt.show()


plt.bar(labels, Mean_Sepal_Width, color='lightgreen')
plt.xlabel('Species')
plt.ylabel('Mean Sepal Width (cm)')
plt.title('Bar Plot of Mean Sepal Width by Species')
plt.show()
```

```python
Mean_Petal_Length = df.groupby('species')['petal_length'].mean()
Mean_Petal_Width = df.groupby('species')['petal_width'].mean()

summary_stats = pd.DataFrame({
    'Mean Petal Length': Mean_Petal_Length,
    'Mean Petal Width': Mean_Petal_Width
}).reset_index()
```

```python
labels = summary_stats['species']
Mean_Petal_Length = summary_stats['Mean Petal Length']
Mean_Petal_Width = summary_stats['Mean Petal Width']

plt.bar(labels, Mean_Petal_Length, color='skyblue')
plt.xlabel('Species')
plt.ylabel('Mean Petal Length (cm)')
plt.title('Bar Plot of Mean Petal Length by Species')
plt.show()


plt.bar(labels, Mean_Petal_Width, color='lightgreen')
plt.xlabel('Species')
plt.ylabel('Mean Petal Width (cm)')
plt.title('Bar Plot of Mean Petal Width by Species')
plt.show()
```
